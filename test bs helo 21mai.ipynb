```python
# Set up
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import numpy as np
from pprint import pprint
#To input random time when scrapping => otherwise detected as a robot
from time import sleep
from random import randint
```

```python
movies=[]
for page in range(1,2): #the last number indicates steps (as we do not want all the films => too much data) 7646,3
    raw_url = 'https://www.allocine.fr/films/alphabetique/?page={}'
    response = requests.get(raw_url.format(str(page)))
    html=response.text
    soup = BeautifulSoup(html,'html.parser')
    movies_all =soup.find_all('li','mdl')
    print(movies_all)
    sleep(randint(1,5))
    #initialize
    date = np.nan
    length = np.nan
    nation = np.nan
    director = []
    actor = []
    genre = []
    for movie in movies_all:
        #title
        title = movie.select_one('a.meta-title-link').get_text()
        #date sortie
        date = movie.select_one('span.date').get_text()
        #duration
        length = movie.select_one('span.spacer').next_sibling.get_text() #next sibling is used to find the duration (otherwise, we just get the / from before)
        length= str(re.sub("[\n]","",length)) #to remove the slash n => manifestation of spaces
        #genres
        info= movie.select_one("meta-body-item meta-body-info")
        print(info)
        all_genres= info.find_all("span",class_=re.compile(r".*==$"))
        print(all_genres)
        if len(all_genres==1):
            genre_1=all_genres[0].get_text()
        elif len(all_genres==2):
            genre_1=all_genres[0].get_text()
            genre_2=all_genres[1].get_text()
        else:
            genre_1=all_genres[0].get_text()
            genre_2=all_genres[1].get_text()
            genre_3=all_genres[2].get_text()
        #Director
        #director = movie.select_one("span",class_=re.compile(r".*blue-link$")).get_text()
        #print(all_genres)
        movies.append({
            'title': title,
            'date':date,
            'duration': length,
            'genre 1':genre_1,
            'genre 2':genre_2,
            'director': director,
        })
        sleep(randint(1,5))
moviedata= pd.DataFrame(movies)
moviedata

#<a class="xXx" href="/films/genre-13005/">Comédie</a>
#<a class="xXx blue-link" #href="/personne/fichepersonne_gen_cpersonne=182821.html">Radu Jude</a>
```
